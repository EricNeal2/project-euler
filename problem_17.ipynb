## [Problem 17](https://projecteuler.net/problem=17): Number letter counts

If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.

First I will create dictionaries of words:

```python
ones_dict = {'0': '',
             '1': 'one',
             '2': 'two',
             '3': 'three',
             '4': 'four',
             '5': 'five',
             '6': 'six',
             '7': 'seven',
             '8': 'eight',
             '9': 'nine'}

tens_dict = {'0': '',
             '1': 'ten',
             '2': 'twenty',
             '3': 'thirty',
             '4': 'forty',
             '5': 'fifty',
             '6': 'sixty',
             '7': 'seventy',
             '8': 'eighty',
             '9': 'ninety'}

teens_dict = {'10': 'ten',
             '11':'eleven',
             '12': 'twelve',
             '13': 'thirteen',
             '14': 'fourteen',
             '15': 'fifteen',
             '16': 'sixteen',
             '17': 'seventeen',
             '18': 'eighteen',
             '19': 'nineteen'}
```

Next I will create a function that returns the length of the word of a number. Because of the way the english language works, there are a lot of special cases like when to use the word 'and', the teens, etc.

```python
def letters_count(num):
    num_str = str(num)
    digit_count = len(num_str)
    
    num_as_words = ''
    
    # HUNDREDS COLUMN
    if len(num_str) > 2:
        num_as_words = num_as_words + ones_dict[num_str[-3]] + 'hundred'
        
        # add the 'and' unless it's an exact hundred
        if num_str[-2] != '0' or num_str[-1] != '0':
            num_as_words = num_as_words + 'and'
    
    # TENS COLUMN
    
    # for numbers with 2 or more digits:
    if len(num_str) > 1:
        
        # if it's a teen, use that dict
        if num_str[-2] == '1':
             num_as_words = num_as_words + teens_dict[num_str[-2:]]
        
        # if it's not a teen, use the normal tens and ones dictionaries
        else:
            num_as_words = num_as_words + tens_dict[num_str[-2]]
            
            # ONES COLUMN
            num_as_words = num_as_words + ones_dict[num_str[-1]]
            
    # ONES COLUMN for single digit numbers
    if len(num_str) == 1:
        num_as_words = num_as_words + ones_dict[num_str[-1]]
    
    
    print(num_as_words)
    
    return len(num_as_words)

print(letters_count(57))
print(letters_count(12))
print(letters_count(658))
print(letters_count(218))
print(letters_count(342))
print(letters_count(115))
print(letters_count(500))
print(letters_count(7))
```

I didn't give my function the ability to handle numbers with 4 digits, so I will calculate the sum for 1-999 and then add on 'onethousand' manually:

```python
total = sum([letters_count(i) for i in range(1000)])
```
